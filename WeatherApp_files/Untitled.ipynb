```python
import csv
```

```python
file = "weatherHacks.csv"

place_list = []
id_list = []

with open(file, 'r') as f:
    reader = csv.reader(f)
    header = next(reader)  # ヘッダーを読み飛ばしたい時

    for row in reader:
        place_list.append(row[0])
        if(row[1] == ""):
            id_list.append('-1')
        else:
            id_list.append(row[1])
        print( row ) 
```
